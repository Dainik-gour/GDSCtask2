# Consider a (5,5,8) shape array, what is the index (x,y,z) of the 100th element?

```python
shape = (5, 5, 8)
element_number = 100


index = element_number - 1


z = index // (shape[0] * shape[1])


remaining_index = index % (shape[0] * shape[1])


x = remaining_index // shape[1]
y = remaining_index % shape[1]

print(f"Index (x, y, z) of the 100th element: ({x}, {y}, {z})")
```

# How to find common values between two arrays? Demonstrate.

```python
#You can find common values between two arrays in Python using various approaches,
# such as list comprehensions, sets, or NumPy for numerical arrays. 
#Here are a few examples demonstrating these methods:

array1 = [1, 2, 3, 4, 5]
array2 = [3, 4, 5, 6, 7]

common_values = [value for value in array1 if value in array2]

print("Common Values:", common_values)
```

```python
#Using Python Sets 
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 7}

common_values = set1.intersection(set2)

print("Common Values:", common_values)
```

```python
#Using NumPy
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([3, 4, 5, 6, 7])


common_values = np.intersect1d(array1, array2)

print("Common Values:", common_values)
```

# Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates.


```python
c_coords = np.random.rand(10, 2)


x = c_coords[:, 0]
y = c_coords[:, 1]

r = np.sqrt(x**2 + y**2)

theta = np.arctan2(y, x)

theta_degrees = np.degrees(theta)

p_coords = np.column_stack((r, theta_degrees))

print("Cartesian Coordinates:")
print(c_coords)

print("\nPolar Coordinates (r, θ in degrees):")
print(p_coords)
```

# Question 4

```python
df = pd.read_csv('GDSC_Recruitment_Form .csv')
```

```python
df.drop_duplicates(subset=['Email Address'], keep='first', inplace=True)
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.sort_values(by='Timestamp', inplace=True)
```

```python

df['BITS ID'] = df['BITS ID'].str.replace('[Ff]|@.*', '', regex=True)
```

```python
df.columns
```

```python
df['Year'] = df['Email Address'].str.extract(r'(\d{4})')
```

```python
df
```

```python
import matplotlib.pyplot as plt

vertical_counts = df['Vertical you wish to apply for'].value_counts()


plt.figure(figsize=(8, 6),dpi=100)

plt.bar(vertical_counts.index, vertical_counts.values, color='#33A1C9')

plt.xlabel('Vertical', fontsize=14)
plt.ylabel('Number of Applicants', fontsize=14)
plt.title('Number of Applicants per Vertical', fontsize=16)

plt.xticks(rotation=45)

plt.gca().set_facecolor('#F0F0F0')

plt.grid(axis='y', linestyle='--', alpha=0.2)

for i, count in enumerate(vertical_counts):
    plt.text(i, count + 1, str(count), ha='center', va='bottom', fontsize=12, color='black')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

year_data = df['Year']

plt.figure(figsize=(5, 6))
n, bins, patches = plt.hist(year_data, bins=4, edgecolor='k', alpha=0.7)

plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Applicants', fontsize=14)
plt.title('Number of Applicants per Years', fontsize=16)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

for i in range(len(patches)):
    plt.text(bins[i] + 0.5, n[i] + 1, str(int(n[i])), ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
df.to_csv('applicants.csv', index=False)
```

```python
verticals = df['Vertical you wish to apply for'].unique()
for vertical in verticals:
    vertical_df = df[df['Vertical you wish to apply for'] == vertical]
    vertical_df.to_csv(f'{vertical}.csv', index=False)
```
